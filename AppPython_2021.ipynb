### factorial() 함수 작성 방법
> 재귀호출과 반복문 사용

````python
def factorial(x):
    if x == 0:
        return 1
    else:
        return x * factorial(x - 1)

print (factorial(5))
````

````python
def factorial(x):
    fac = 1
    for i in range(1,x+1):
        fac *= i
    return fac

print (factorial(5))
````

* 자료에서 특정한 조건의 값을 찾아 합을 구한다.

````python
# 다음 리스트에서 50점 이상의 점수들의 총합을 구하시오.
marks=[90,25,67,45,80]

total = 0
i = 0
higher_cnt = 0
for i in range(len(marks)):
    #print(f'i = {i}')
    if marks[i] >= 50:
        higher_cnt += 1
        total = total + marks[i]

print ("50점 이상의 점수들의 총합 = %d" % total)
print ("50점 이상의 점수들의 평균 = %d" % (total/higher_cnt))
````

* 소수(prime number) 구하기

````python
for n in range(2,10):
  for x in range(2,n):
    if n % x == 0:
      print (n, 'equals', x, '*', n/x)
      break
  else:
    # loop fell through without finding a factor
    print (n, 'is prime')
````

###for - else

num = 8 에서 반복루프를 빠져나오기때문에, else문이 실행이 되지 않고
**list contains numbers greater than 5** 이 출력 됨

````python
lst = [1,2,3,4,8]
for num in lst:
      if num >5:
          print ("list contains numbers greater than 5")
          break
else:
      print("list contains numbers less than 5");
````

for loop가 break없이 빠져나왔으므로 **list contains numbers less than 5** 가 출력이 된다.

````python
lst = [1,2,3,4]
for num in lst:
      if num >5:
          print ("list contains numbers greater than 5")
          break
else:
      print("list contains numbers less than 5");
````

## (실습) for 문

소수(prime number) 구하기

````python
for n in range(2,10):
  for x in range(2,n):
    if n % x == 0:
      print (n, 'equals', x, '*', n/x)
      break
  else:
    # loop fell through without finding a factor
    print (n, 'is prime')
````

## 내장함수  zip() 
여러 개의 순회 가능한(iterable) 객체를 인자로 받고, 각 객체가 담고 있는 원소를 터플의 형태로 차례로 접근할 수 있는 반복자(iterator)를 반환

````python
grades = ["High", "Medium", "Low"]
values = [0.75, 0.50, 0.25]

print (*zip(grades, values))

for grade, value in zip(grades, values):
    print("{}% {}".format(value * 100, grade))
````
